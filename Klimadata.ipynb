```python
from arcgis.gis import GIS
from arcgis.features import FeatureLayer

# Connect to GIS
gis = GIS()

# URL til et lag i Klimaatlas
url = "https://klimaatlas-dmidata.opendata.arcgis.com/datasets/klimaatlas_dmidata_openapi/arctic-temperatures"

# Fetch layer
layer = FeatureLayer(url)

# Fetch data
data = layer.query(where="1=1", out_fields="*")

# Convert to a GeoDataFrame
df = data.sdf
print(df.head())
```

```python
import requests

# URL til Klimaatlas data via ArcGIS
url = "https://klimaatlas-dmidata.opendata.arcgis.com/datasets/klimaatlas_dmidata_openapi/arctic-temperatures"

# Hent data
response = requests.get(url)

# Tjek statuskode
if response.status_code == 200:
    data = response.json()  # Assuming the data is in JSON format
    print(data)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```
